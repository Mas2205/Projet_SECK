```python
import gensim
import os 
```

1) Chargement des données textuelles

```python
path='Data/'
l=[]
for rep in os.listdir(path):
    for fichier in os.listdir(path+rep):
        print(rep,fichier)
        with open(path+rep+'/'+fichier,'r',encoding="utf8") as f:
            l.append(f.read().replace('\n\n','.').replace('\n',' ').strip())
```

2) Le nettoyage

Tokens (dont ponctuation de fin de phrase) / stop words 



```python
import re


def tok_me(texte):
    # Je modifie la chaine de caractere s'il y a une ponctuation
    texte=texte.replace('.',' . ').replace('!', ' ! ').replace('?', ' ? ')
    p="([aA-zZéèàùîêâûôçïëœ.!?-]{1,})"
    texte=' '.join(re.findall(p,texte))
    #Je renvoi une liste de token
    return texte.split()

with open('stop-w_fr.txt',"r",encoding='utf8') as f:
    stop=f.read().split('\n')
stop.remove('haut')
stop.remove('pas')
stop.remove('ne')
stop.remove('n')
stop.remove('ni')
stop.remove('peu')
stop.remove('bon')
stop.remove('très')
stop.remove('tres')
stop.remove('même')
stop.remove('fait')
stop.remove('début')
stop.remove('mais')
stop.remove('dès')
stop.remove('premier')
stop.remove('première')
stop.remove('plus')
stop.remove('aussi')
stop.remove('bien')
stop.remove('assez')
stop.remove('retour')
stop.remove('tout')
stop.remove('rien')

def stop_me(liste_token):
    final=[]
    for token in liste_token:
        if token in stop:
            continue
        final.append(token)
    return final



stop_me(tok_me(l[0]))
```

3) Decoupage en phrase

liste=[phrase 1, phrase 2, phrase3, phrase N] (où N est le nombre de phrase de tous les textes)
et phrase 1 = [mot1,mot2, ...,motn] (où n est le nombre de mot de la phrase 1)

```python
l[0]
```

```python
def cut_sent(tokens):
    final=[]
    tmp=[]
    for tok in tokens:
        if tok in [".","!","?"]:
            if tmp != []:
                final.append(tmp)
            tmp=[]
        else:
            tmp.append(tok)
    if tmp != []:
        final.append(tmp)
    return final

cut_sent(stop_me(tok_me(l[0])))
```

Generalisation sur tous les livres

```python
from tqdm import tqdm

final=[]
for livre in tqdm(l):
    final=final+cut_sent(stop_me(tok_me(livre)))
    
```

4) Construire un modèle type Word2Vec avec gensim

```python
from gensim.models import Word2Vec

model = Word2Vec(sentences=final, vector_size=50, window=5)
model.save("word2vec.model")
```

Text feat

```python
import numpy as np

def sent_enc(chaine):
    summy=np.zeros(50)
    i=0
    for token in stop_me(tok_me(chaine)):
        try:
            summy=model.wv.get_vector(token)+summy
            i=i+1
        except:
            continue
    return summy / i


text1 = "Olympique de Marseille peut se targuer d’avoir pu compter sur Didier Drogba, Steve Mandanda, Dimitri Payet ou Franck Ribéry"

sent_enc(text1)
```
